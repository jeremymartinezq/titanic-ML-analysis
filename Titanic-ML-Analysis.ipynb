```python
# Titanic ML Seaborn Analysis

# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set style
sns.set_style('whitegrid')

# Load dataset
titanic = sns.load_dataset("titanic")

# Create correlation matrix in the correct order
variables = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']
correlation_matrix = titanic[variables].corr()

# Create a figure for the table
plt.figure(figsize=(12, 3))
plt.axis('off')

# Create table with the first 5 rows
table_data = titanic.head()
table = plt.table(cellText=table_data.values,
                 colLabels=table_data.columns,
                 cellLoc='center',
                 loc='center',
                 colColours=['#f2f2f2']*len(table_data.columns))

# Adjust table properties
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1.2, 1.5)

plt.title("First Five Rows of Titanic Dataset", pad=20)
plt.show()

# Joint Plot
g = sns.jointplot(data=titanic, x="fare", y="age", kind="scatter")
g.fig.suptitle("Joint Distribution of Fare and Age", y=1.02)
plt.show()

# Distribution Plot (using distplot for exact match)
plt.figure(figsize=(8, 6))
sns.distplot(titanic['fare'].dropna())
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x="class", y="age", data=titanic)
plt.title("Age Distribution by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()

# Swarmplot
plt.figure(figsize=(8, 6))
sns.swarmplot(x="class", y="age", data=titanic)
plt.title("Age Distribution by Class (Swarm Plot)")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()

# Count Plot
plt.figure(figsize=(8, 6))
sns.countplot(x="sex", data=titanic)
plt.title("Passenger Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Pair Plot
g = sns.pairplot(titanic[['survived', 'age', 'fare', 'sex']], hue='sex')
g.fig.suptitle("Pair Plot of Titanic Variables", y=1.02)
plt.show()

# Create heatmap with correct order
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, 
            annot=True,
            cmap='RdBu_r',
            center=0,
            vmin=-1,
            vmax=1,
            square=True,
            fmt='.2f')
plt.title("Correlation Heatmap of Titanic Variables")
plt.tight_layout()
plt.show()

# FacetGrid
g = sns.FacetGrid(titanic, col="sex")
g.map(sns.histplot, "age")
g.fig.suptitle("Age Distribution by Gender", y=1.02)
plt.show()


```
